```python
%matplotlib inline

from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(14,6), dpi=80)
plt.subplot(1, 1, 1)

cosColor = 'blue'
sinColor = 'red'

X = np.linspace(-np.pi, np.pi, 256, endpoint=True)
C, S = np.cos(X), np.sin(X)

plt.plot(X, C, color=cosColor, label='cosine', linewidth=2.5, linestyle='-')
plt.plot(X, S, color=sinColor,  label='sine', linewidth=2.5, linestyle='-')

plt.legend(loc="upper left")

axes = plt.gca()
axes.spines['right'].set_color('none')
axes.spines['top'].set_color('none')
axes.xaxis.set_ticks_position('bottom')
axes.spines['bottom'].set_position(('data', 0))
axes.yaxis.set_ticks_position('left')
axes.spines['left'].set_position(('data', 0))

plt.xlim(X.min() * 1.1, X.max() * 1.1)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi],
          [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])

plt.ylim(C.min() * 1.1, C.max() * 1.1)
plt.yticks([-1, -0.5, 0, 0.5, +1],
          [r'$-1$', r'$-\frac{1}{2}$', r'$0$', r'$\frac{1}{2}$', r'$+1$'])

px = 2 * np.pi / 3
plt.plot([px, px], [0, np.cos(px)], color=cosColor, linewidth=1.0, linestyle='--')
plt.scatter([px], [np.cos(px)], 50, color=cosColor)

plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$',
            xy=[px, np.cos(px)], xycoords='data',
            xytext=(-90, -50), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=.2'))

plt.plot([px, px], [0, np.sin(px)], color=sinColor, linewidth=1.0, linestyle='--')
plt.scatter([px], [np.sin(px)], 50, color=sinColor)

plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$',
            xy=[px, np.sin(px)], xycoords='data',
            xytext=(30, 20), textcoords='offset points', fontsize=16,
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3, rad=.2'))

for label in axes.get_xticklabels() + axes.get_yticklabels():
    label.set_fontsize(16)
    label.set_bbox(dict(facecolor='white', edgecolor='none', alpha=0.30))
    

#plt.savefig('sin_cos_with_cfg.png', dpi=120)

plt.show()
```
